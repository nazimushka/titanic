```python
df=train
```

```python
df
```

```python
#удаление ненужных колонок
useless_columns = ['Name', 'Ticket', 'Cabin']
tc_df = df.drop(columns=useless_columns)
```

```python
tc_df
```

```python
# метод .isnull() выдает логический массив, где пропуски обозначены как True
# метод .sum() по умолчанию суммирует эти True или единицы по столбцам
tc_df.isnull().sum()
```

```python
# заполним пропуски в столбце Age медианным значением
tc_df.Age.fillna(tc_df.Age.median(), inplace = True)
# два пропущенных значения в столбце Embarked заполним портом Southhampton
tc_df.Embarked.fillna('S', inplace = True)
# проверим результат (найдем общее количество пропусков сначала по столбцам, затем по строкам)
tc_df.isnull().sum().sum()
```

```python
sns.catplot(data = tc_df,y='Survived',x='Sex',col='Pclass', kind='bar', saturation=0.5) #Диаграмма 1
```

# *Вывод*: Женщин спаслось гораздо больше, чем мужчин в каждом из классов. Это говорит как минимум о порядочности большинства пассажиров - мужчин

```python
sns.catplot(data = df,hue = 'Survived', x = 'Pclass', kind='count', saturation=0.5)
# Диаграмма 2
```
